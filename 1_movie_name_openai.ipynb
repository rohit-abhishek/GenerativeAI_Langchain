````python
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate
````

````python
system_prompt = SystemMessagePromptTemplate.from_template("""You are moviemaker and known for creating creative movie title.""")
human_prompt = HumanMessagePromptTemplate.from_template("""You are tasked to read the premise of a movie: 
                                                        {article}. 
                                                        Come up with a name for this movie. 
                                                        Name should be in catchy in English or Hindi language. Provide meaning of the title as well""",
                                                        input_variables=['article'])
````

````python
human_prompt.format(article=article)
````

````python
from langchain_core.prompts import ChatPromptTemplate

# create the first prompt 
first_prompt = ChatPromptTemplate.from_messages([system_prompt, human_prompt])
````

````python
first_prompt.format(article="Testing")
````

# Using LCEL - Langchain Expression Langauge 
We will be using chain to perform - 
- Formatting 
- LLM Generation 
- Output 

The object is composed of value which appears separated by pipe character. What is the on the left of pipe is the input to what is on the right side of pipe 

````python
chain_one = (
    {"article": lambda x: x["article"]}
    | first_prompt
    | creative_llm
    | {"Movie Name" : lambda x : x.content}
)
````

````python
type(chain_one)
````

````python
chain_one.invoke({"article": article})
````

````python
chain_two = (
    {"article": lambda x: x["article"]}
    | first_prompt
    | boring_llm
    | {"Movie Name" : lambda x : x.content}
)
````

````python
chain_two.invoke({"article" : article})
````
